### V16 - Oszillometrische Blutdruckmessung (iOBP)

In diesem Versuch haben wir mit zwei Blutdruckmessgeräten (Beurer und Nais) Referenzmessungen durchgeführt, sowie  mit einem Prototypischen-Gerät, für dessen Messungen ein Algorithmus entwickelt wurde, um den Refenzmessungen möglichst nahe zu kommen.

Dieser Algorithmus findet sich in der Datei `BlutdruckMesser.py`. Der Algorithmus hat insgesamt 9 einstellbare Parameter. Da eine manuelle Einstellung dieser zu aufwendig wäre, haben wir eine Biobliothek namens Optuna (https://optuna.org) dafür verwendet. In der Datei `optimzer.py` wurden dann die Parameter optimiert und die Ausgabe haben wir in den `results parameter.txt` gespeichert.

In diesem Jupyter Notebook werden wir nun alle Messungen analysieren, plotten und die Abweichung zur Referenz berechnen.

### Messungen optimiert auf Kombination von Beurer und NAIS

```python
from BlutdruckMesser import Signal, alogrithmus
from Messungen.Infos import MEASUREMENT_INFORMATION

abweichung_beurer_sys = 0
abweichung_beurer_dia = 0
abweichung_nais_sys = 0
abweichung_nais_dia = 0

counter_beurer = 0
counter_nais = 0
for key, value in MEASUREMENT_INFORMATION.items():
    messnummer = key
    b_sys, b_dia = value["Beurer"]
    n_sys, n_dia = value["NAIS"]
    b_sys = float(b_sys) if b_sys is not None else None
    b_dia = float(b_dia) if b_dia is not None else None
    n_sys = float(n_sys) if n_sys is not None else None
    n_dia = float(n_dia) if n_dia is not None else None
    print(f"Analysieren der Messung: {messnummer}")
    calc_sys, calc_dia, signal = alogrithmus(
        messnummer=messnummer,
        begin_index=1,
        high_N=5,
        low_N=5,
        border_f=1.4727302354599732,
        use_absolute_value=True,
        peaks_distance=165,
        window_size=5.657079422588785,
        dia_treshhold=0.794095733069117,
        sys_trashhold=0.8626074091583769
    )
    signal.plot_data(title=f"Optimized for Beurer and NAIS: {messnummer}", filename=f"combined_{messnummer}", save_plot=True)
    print(f"\nBerechnete Werte: \t\t {calc_sys:.2f}\t/\t{calc_dia:.2f}\tmmHg")
    if b_sys is not None and b_dia is not None:
        a_sys = abs(calc_sys - b_sys)
        a_dia = abs(calc_dia - b_dia)
        print(f"\nGround Truth Beurer:\t {b_sys}\t/\t{b_dia}\tmmHg")
        print(f"Abweichung Beurer:\t\t {a_sys:.2f}\t/\t{a_dia:.2f}\tmmHg")
        abweichung_beurer_sys += a_sys
        abweichung_beurer_dia += a_dia
        counter_beurer += 1
    else:
        print(f"Keine Messungen mit Beurer vorhanden")

    if n_sys is not None and n_dia is not None:
        a_sys = abs(calc_sys - n_sys)
        a_dia = abs(calc_dia - n_dia)
        print(f"Ground Truth NAIS:\t\t {n_sys}\t/\t{n_dia}\tmmHg")
        print(f"Abweichung NAIS:\t\t {a_sys:.2f}\t/\t{a_dia:.2f}\tmmHg")
        abweichung_nais_sys += a_sys
        abweichung_nais_dia += a_dia
        counter_nais += 1
    else:
        print("Keine Messungen mit NAIS vorhanden")

    print("\n\n")

print("Allgemeine Analyse:")
print(f"Abweichung Beurer: {abweichung_beurer_sys/counter_beurer}/{abweichung_beurer_dia/counter_beurer}mmHg")
print(f"Abweichung NAIS: {abweichung_nais_sys/counter_nais}/{abweichung_nais_dia/counter_nais}mmHg")
```
